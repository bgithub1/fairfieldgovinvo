### Fairfield Polling Places

Create a csv file of polling locations by street and school using info from:
* https://www.fairfieldct.org/filestorage/10726/12538/12540/12463/12467/RTM_District_Street_Listing_including_Polling_Locations_-_10-05-2015.pdf

The polling.txt file was created by downloading the above url manually (had a problem using requests), and then using Apache pdfbox to extract the text from the pdf

```python
import pandas as pd
import numpy as np
import requests
import re
```

```python
# Extract text from pdf using Apache pdfbox jar (you need java installed)
!bash pdfjar_extract_text.sh data polling
```

```python
txt = open('data/polling.txt','r').readlines()
sn = 'STREET NAME SIDE'
header = 'STREET NAME SIDE RANGE D/P LO/D SP/DPOLLING PLACE'
start_loc = None
for i in range(len(txt)):
    t = txt[i]
    if sn in t:
        start_loc=i
        break

lines = []
locs = []
streets = []
schools = []
loc_re = ' [0-9]{1,}[^0-9][0-9]{1,}[ ][0-9]{1,}[^0-9][0-9]{1,}'
for t in txt[start_loc:]:
    if 'STREET' not in t:
        lines.append(t)
for t in lines:
    try:
        parts = re.split(loc_re,t.strip())
        loc = re.findall(loc_re,t.strip())[0]
        schools.append(parts[1])
        streets.append(parts[0])
        locs.append(loc)
    except Exception as e:
        print(e)
        print(t)
        continue

df_poll = pd.DataFrame({'street':streets,'school':schools,'loc':locs})
df_poll
```

```python
df_poll.to_csv('polling_places.csv',index=False)
```
