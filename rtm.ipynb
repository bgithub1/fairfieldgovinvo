### Composition of Fairfield RTM as of 2020-10-27

```python
import pandas as pd
import numpy as np
```

```python
dfs = pd.read_csv(f'data/rtm.txt')
pp = lambda v:np.nan if len(v)<3 else v[2:]
dfs['email'] = dfs.party.apply(pp)
dfs['party'] = dfs.party.str.slice(0,1)
dfs.name = dfs.name.str.strip()
dfs.email = dfs.email.str.strip()
dfs.level = dfs.level.str.strip()
dfs['last_name'] = dfs.name.apply(lambda v:v.split(' ')[-1])
dfs['first_name'] = dfs.name.apply(lambda v:' '.join(v.split(' ')[0:-1]))
dfs = dfs[[c for c in dfs.columns.values if c!='name']]
dfs.to_csv('data/rtm.csv',index=False)
```
